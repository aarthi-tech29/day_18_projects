### Project 2: Visualizing a 3D Gradient
#### Objective: Plot a 3D surface for a function like f(x,y)=x^2+y^2 and a vector showing its gradient at a specific point. 
#### Technique: Use Matplotlib's mplot3d toolkit to plot the surface and a quiver plot to show the gradient vector. 

##### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

##### Step 2: Create the grid of x and y values

```python
x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2  # Function values

# np.linspace(-3, 3, 30) → creates 30 evenly spaced numbers between -3 and 3.
# np.meshgrid(x, y) → turns x and y into a grid of coordinate
# Z = X**2 + Y**2 → This gives the height (Z) of the surface at each (X, Y).
```

##### Step 3: Choose a specific point to show the gradient

```python
x0, y0 = 1, 1
z0 = x0**2 + y0**2
grad_x = 2 * x0
grad_y = 2 * y0

# We pick the point (1, 1).
# Compute its height: z0​ = 1^2 + 1^2 = 2
# The gradient of f(x,y) is [2x,2y]
# So, ∇f(1,1)=[2,2]
# This gradient tells the direction of steepest increase.
```

##### Step 4: Create the 3D plot

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d') # add_subplot(nrows, ncols, index) one big plot, 111 means 1 → number of rows of plots, 1 → number of columns of plots, 1 → this is the first (and only) plot

# plt.figure() → starts a new figure window.
# projection='3d' → tells matplotlib want a 3D plot.

# Plot the surface
ax.plot_surface(X, Y, Z, alpha=0.6, cmap='viridis')

# plot_surface() → draws the 3D surface.
# alpha=0.6 → makes it slightly transparent .
# cmap='viridis' → gives it a green-blue color gradient.

# Plot the gradient vector
ax.quiver(x0, y0, z0, grad_x, grad_y, 0, color='red', length=1.0, normalize=True, zorder=5)

# ax.quiver() → draws a 3D arrow-draws the gradient vector at (x0, y0, z0)
# Starts at (x0, y0, z0) → the starting point on the arrow
# Direction is (grad_x, grad_y, 0) → the gradient vector (arrow shows direction of increase)
# length=1.0 → controls the arrow size.
# normalize=True → scales the arrow so it fits nicely
# red arrow - Gradient vector (points uphill) It shows the direction where f(x,y) is increasing(maximum) the most rapidly

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x,y)')
ax.set_title('3D Surface with Gradient Vector')
plt.show()
```
