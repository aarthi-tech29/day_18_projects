### Project 2: Simple Gradient Descent for Linear Regression
#### Objective: Implement a basic linear regression model from scratch using gradient descent

##### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Step 2: Generate Data

```python
np.random.seed(0)
x = np.linspace(0, 10, 50)
y = 2.5 * x + 1.0 + np.random.randn(50) * 2

# np.random.seed(0) → fixes the random numbers so the output is the same every time
# np.linspace(0, 10, 50) → creates 50 evenly spaced x-values from 0 to 10
# y = 2.5 * x + 1.0 + noise → builds a linear relationship between x and y,but adds random noise (np.random.randn(50) * 2) to make it look real
# Noise means random variations or errors in your data
```

##### Step 3: Define the Loss Function

```python
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

# Mean Squared Error (MSE)
# It measures how far off the predictions are from the real data.
# The smaller this number, the better the line fits the data
# MSE - Measures how wrong predictions are
```

##### Step 4: Define Gradients

```python
def compute_gradients(x, y, m, b):
    N = len(x)
    y_pred = m * x + b
    dm = (-2/N) * np.sum(x * (y - y_pred))
    db = (-2/N) * np.sum(y - y_pred)
    return dm, db

# This function finds how much we need to adjust m (slope) and b (intercept) each step
# dm: change needed for slope
# db: change needed for intercept
# Gradient - Direction and size of change needed
```

##### Step 5: Initialize parameters

```python
m = np.random.randn()   # start with random slope
b = np.random.randn()   # start with random intercept
learning_rate = 0.01
epochs = 1000

# start with random guesses for m and b,then improve them little by little using gradient descent
# Learning rate - Controls how big each update step is
# epochs - Number of times we repeat the learning loop
# m,b -Slope and intercept of our best-fit line

# Lists to store loss values for plotting
loss_history = []
```

##### Step 6: Gradient Descent Loop

```python
for epoch in range(epochs):
    y_pred = m * x + b
    loss = mse(y, y_pred)
    loss_history.append(loss)   # record loss each epoch

    dm, db = compute_gradients(x, y, m, b)
    # Update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db

# This repeats 1000 times (epochs)
# Predict y_pred using current m and b
# Calculate loss (how wrong we are)
# Compute gradients (dm, db) → how to fix m and b
# Update m and b a little bit in the direction that reduces loss

    if epoch % 100 == 0:
        print(f"Epoch {epoch}: m={m:.3f}, b={b:.3f}, Loss={loss:.3f}")

        # The current epoch number
        # The current values of m and b
        # The loss value
        # The loss getting smaller → model is improving

```

##### Step 7: Final Model Plot

```python
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, m*x + b, color='red', label='Best-fit line (Gradient Descent)')
plt.title('Linear Regression using Gradient Descent')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Blue points → actual data.
# Red line → best-fit line after gradient descent learns optimal m and b
```

##### Step 8: Loss Curve Plot

```python
plt.figure(figsize=(8,5))
plt.plot(loss_history, color='green', linewidth=2)
plt.title('Loss Decreasing Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid(True)
plt.show()

print("Final model parameters and loss:")
print("Slope (m) ≈", round(m, 4))
print("Intercept (b) ≈", round(b, 4))
print ("Minimum of loss ≈", round(loss,4))
```
