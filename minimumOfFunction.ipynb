### Project 1: Finding the Minimum of a Function 
#### Objective: Use a simple gradient descent implementation to find the minimum of a single-variable function like f(x)=x^4−3x^3+2. 
#### Technique: Start at a random x, calculate the derivative, and iteratively update x until the derivative is close to zero. 

##### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Step 2: Define the function and its derivative

```python
def f(x):
    return x**4 - 3*x**3 + 2
def df(x):
    return 4*x**3 - 9*x**2

# df(x) is the derivative of f(x).
# The derivative tells us the slope at any point x.
# Gradient descent uses this slope to know which direction to move x to reduce the function value.
```

##### Step 3: Initialize variables

```python
np.random.seed(0)
x = np.random.randn()  # Start at a random point
learning_rate = 0.01
tolerance = 1e-6
max_iters = 10000

# x = random starting point (can be anywhere on the x-axis).
# learning_rate = how big a step we take in each iteration. Too big → might overshoot minimum, too small → takes too long.
# tolerance = stop when the slope (derivative) is smaller than this, meaning we’re very close to a flat point.
# max_iters = maximum number of iterations to prevent infinite loops.
# 1e-6 - This is because if the slope is extremely close to 0, it means we are almost at a minimum
# If grad = 0.0000005 → |grad| < 1e-6 → stop
# If grad = 0.01 → |grad| > 1e-6 → keep updating x
```

##### Step 4: Gradient Descent Loop

```python
for i in range(max_iters):
    grad = df(x)
    if abs(grad) < tolerance:
        break
    x = x - learning_rate * grad

# grad = df(x) → calculate the slope at current x.
# abs(grad) converts it to a positive number so we can compare it to tolerance
# if abs(grad) < tolerance: break → if slope is nearly zero, stop (we’ve reached a minimum).
# x = x - learning_rate * grad → move x opposite to the slope:
# If slope is positive → move left (decrease x).
# If slope is negative → move right (increase x).
# This repeats until the slope is very small or the maximum number of iterations is reached.



```

```python
x_vals = np.linspace(-1, 3, 400)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='f(x)')
plt.scatter(x, f(x), color='red', label='Minimum found')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Finding Minimum')
plt.legend()
plt.grid(True)
plt.show()

print("Minimum at x =", round(x,4))
print("Minimum value f(x) =", round(f(x),4))

# Shows the x where the minimum occurs.
# Shows the function value f(x) at that point.
```
