### Project 1: Visualizing a Derivative 
#### Objective: Plot a function and its derivative to get an intuitive feel for the concept. 

##### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Step 2: Define the Function and Its Derivative

```python
def f(x):
    return x**2

def f_prime(x):
    return 2*x

# f(x)=x^2→ this is our main function
# f′(x)=2x→ this gives the slope (derivative) of f(x)
# derivative tells us how fast the function changes at any point
```

##### Step 3: Create Data Points for X and Y

```python
x = np.linspace(-5, 5, 100)
y = f(x)

# np.linspace(-5, 5, 100) creates 100 evenly spaced points between -5 and 5.
# calculate y = f(x) → that gives us y=x^2 values
```

##### Step 4: Calculate outputs

```python
y = f(x)         # Original function values
y_prime = f_prime(x)  # Derivative values
```

##### Step 5: Plot both on the same graph

```python
plt.figure(figsize=(8,6))
plt.plot(x, y, label='f(x) = x²', color='blue', linewidth=2)
plt.plot(x, y_prime, label="f'(x) = 2x", color='red', linestyle='--', linewidth=2)

# Add grid, legend, and labels
plt.title("Function and Its Derivative", fontsize=14)
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.8)  # x-axis
plt.axvline(0, color='black', linewidth=0.8)  # y-axis
plt.legend()
plt.grid(True)
plt.show()
```

### Visualizing a Derivatives with tangent lines

##### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Step 2: Define the Function and Its Derivative

```python
def f(x):
    return x**2

def f_prime(x):
    return 2*x

# f(x)=x^2→ this is our main function
# f′(x)=2x→ this gives the slope (derivative) of f(x)
# derivative tells us how fast the function changes at any point
```

##### Step 3: Create Data Points for X and Y

```python
x = np.linspace(-5, 5, 200)
y = f(x)
# np.linspace(-5, 5, 100) creates 100 evenly spaced points between -5 and 5.
# calculate y = f(x) → that gives us y=x^2 values
```

##### Step 4: Choose Points for Tangents

```python
points = [-3, 0, 3]

# x = -3, x = 0, and x = 3 —to show how the slope changes from negative → zero → positive.
```

##### Step 5: Plot the Main Function

##### Step 6: Draw Tangent Lines

```python
plt.figure(figsize=(9, 6))
plt.plot(x, y, label='f(x) = x²', color='blue', linewidth=2)

# Makes a figure of size 9×6 inches.
# Plots the smooth blue curve for f(x)=x^2
for a in points:
    slope = f_prime(a)      # slope at that point
    y_a = f(a)              # y-value of the function
    tangent = slope * (x - a) + y_a  # formula for tangent line

    # tangent line formula - y=m(x−a)+f(a)
    # m=f′(a) → slope (from derivative)
    # (𝑎,𝑓(𝑎)) → the point of contact

    plt.plot(x, tangent, '--', label=f"Tangent at x={a} (slope={slope})")

    plt.scatter(a, y_a, color='black', zorder=5)
    plt.text(a, y_a + 1, f"x={a}", ha='center', fontsize=10)

    # Marks each tangent point with a black dot.
    # Adds a label (x = -3, 0, or 3) slightly above it
    # zorder=5 The black dots (tangent points) appear on top of the lines —not hidden behind them.
    # “scatter” many data points across the graph.

plt.title("f(x) = x² with Tangent Lines Showing the Derivative", fontsize=14)
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.legend()
plt.grid(True)
plt.show()
```
